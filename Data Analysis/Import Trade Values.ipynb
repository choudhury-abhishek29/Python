```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
imports = pd.read_csv('Imports_WITS.csv')
coutry_codes=pd.read_excel('country_code.xlsx')
a_1995 = pd.read_excel('analysis_1995.xlsx', sheet_name='1995')
a_2000 = pd.read_excel('analysis_2000.xlsx', sheet_name='2000')
a_2005 = pd.read_excel('analysis_2005.xlsx', sheet_name='2005')
a_2010 = pd.read_excel('analysis_2010.xlsx', sheet_name='2010')
a_2015 = pd.read_excel('analysis_2015.xlsx', sheet_name='2015')
a_2017 = pd.read_excel('analysis_2017.xlsx', sheet_name='2017')
```

```python
a_1995['trade_value'] = a_1995['Agree']
a_2000['trade_value'] = a_2000['Agree']
a_2005['trade_value'] = a_2005['Agree']
a_2010['trade_value'] = a_2010['Agree']
a_2015['trade_value'] = a_2015['Agree']
a_2017['trade_value'] = a_2017['Agree']
```

```python
c_code = coutry_codes['Numeric']
c_num = coutry_codes['3_letter']
rename_df2 = pd.DataFrame({'c_code':c_code, 'c_num':c_num})
iso_dict = rename_df2.set_index('c_num').to_dict()['c_code']
```

```python
imports_1 = imports.replace({'ReporterISO3': iso_dict, 'PartnerISO3': iso_dict})
imports_1.head(2)
```

```python
imports_1 = imports_1[imports_1['PartnerISO3']!='ANT']
```

```python
ReporterISO3 = imports_1.ReporterISO3.values
PartnerISO3 = imports_1.PartnerISO3.values
merged_list = list(zip(PartnerISO3, ReporterISO3))
imports_1['pairs'] = merged_list
imports_1['pairs'] = (imports_1['pairs'].apply(str)).values
imports_1.rename(columns={'TradeValue in 1000 USD':'trade_value'},inplace=True)
imports_1.head(5)
```

```python
pairs_95 = imports_1[imports_1['Year']==1995]['pairs']
trade_value_95 = imports_1[imports_1['Year']==1995]['trade_value']
trade_value_df95 = pd.DataFrame({'pairs':pairs_95, 'trade_value':trade_value_95})
trade_value_dict95 = trade_value_df95.set_index('pairs').to_dict()['trade_value']

a_1995 = a_1995.replace({'trade_value': trade_value_dict95})
a_1995.head(5)
```

```python
pairs_2000 = imports_1[imports_1['Year']==2000]['pairs']
trade_value_2000 = imports_1[imports_1['Year']==2000]['trade_value']
trade_value_df2000 = pd.DataFrame({'pairs':pairs_2000, 'trade_value':trade_value_2000})
trade_value_dict2000 = trade_value_df2000.set_index('pairs').to_dict()['trade_value']

a_2000 = a_2000.replace({'trade_value': trade_value_dict2000})
a_2000.head(5)
```
